```python
import numpy as np
import pandas as pd
```

Read file

```python
file = open("/content/drive/MyDrive/AnnaK.txt", "rt")
data = file.read()
data = data.replace("-", " ")
data = data.replace(".", " ")
data = data.replace(":", " ")
data = data.replace(";", " ")
data = data.replace(",", " ")
data = data.replace("!", " ")
data = data.replace("(", " ")
data = data.replace(")", " ")
data = data.replace("?", " ")
data = data.replace("'", " ")
data = data.replace("_", " ")
```

```python
print(len(data))
```

Remove any words less than three characters

```python
data = ' '.join([w for w in data.split() if len(w)>3])
```

```python
data[:200]
```

```python
words = data.split()
print(len(words))
```

Remove special characters from text file

```python
pure_words = [i.strip() for i in words if i.strip()]
print(len(pure_words))
```

Convert words to lower case

```python
pure_words = [i.lower() for i in words]
print(pure_words[:200])
```

Count words used in text

```python
# using naive method to remove duplicated from list  
list_word = [] 
for i in pure_words: 
    if i not in list_word: 
        list_word.append(i) 
print(list_word)
```

```python
print(len(list_word))
```

Count the occurance of each word in the text file

```python
# Iterate over each word in line
d = dict()

for word in pure_words:
  # Check if the word is already in dictionary
  if word in d:
    # Increment count of word by 1
    d[word] = d[word] + 1
  else:
    # Add the word to dictionary with count 1
    d[word] = 1
```

```python
# Print the contents of dictionary
for key in list(d.keys()):
  print(key, ":", d[key])
```

Check which words occur only once in text

```python
only_once = []

for key in list(d.keys()):
  if d[key] == 1:
    only_once.append(key)
print(only_once)
```

```python
print(len(only_once))
```
